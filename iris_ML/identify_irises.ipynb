```python
from sklearn import datasets # type: ignore
import seaborn as sns # type: ignore
sns.set_theme()
iris = datasets.load_iris()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_data = iris.data  # Try to convert the Array into the Columns of tables
y_target = iris.target
```

```python
columns = ['Sepal Length' , 'Sepal Width'  , 'Petal Length' , 'Petal Width' ,  'Class' ]
columns_specs = [ (0,3),(4,7), (8,11) , (12, 15), (16,32) ]
data_frame = pd.read_fwf("iris/iris.data", colspecs=columns_specs, names= columns)
data_frame['Class'] = data_frame['Class'].str.replace('Iris-', '')
print(data_frame)
```

```python
sns.pairplot(data_frame , hue="Class")
```

```python
categories = data_frame['Class'].value_counts() # Basically a Dict of different value and its count
data = [categories['setosa'] , categories['versicolor'] , categories['virginica']]
keys = ['setosa', 'versicolor' , 'virginica']
plt.pie(data,labels = keys , autopct='%1.1f%%')
plt.title("Distribution of Data Class out of 150")
plt.show()
```
