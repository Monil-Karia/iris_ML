```python
# Setting the Co-oefficeint 
w  = np.random.rand(2,1) # 6 random Digits Array
b = 0
```

```python
def cost_function(X ,Y , w ,b):
    m = X.shape[0]
    cost = 0
    for i in range(m):
        print("Mera kaam Khatam")
        f_wb = np.dot(X[i], w) + b
        cost += (f_wb - Y[i]) ** 2
    total_cost = cost / (2 * m)
    print(total_cost)
    return total_cost
```

```python
def gradient_descent(X , Y , w , b):
    m = X.shape[0]
    dj_dw = 0
    dj_db = 0
    print(X)
    for i in range(m):
        f_wb  = np.dot(X[i] , w) + b
        print("Yaha Error aa gaya?")
        dj_dw_i = (f_wb  - Y[i]) * X[i]
        dj_db_i = (f_wb  - Y[i])
        dj_db += dj_db_i
        dj_dw += dj_dw_i

    dj_dw = dj_dw/m
    dj_db = dj_db/m
    return dj_db , dj_dw
```

```python
def whole_gradient_descent(X , Y , w , b, cost_function , gradient_function):
    J_history = []
    p_history = []
    alpha = 0.003
    iterations = 10000
    for i in range(iterations):
        dj_dw , dj_db = gradient_function(X , Y , w , b)

        b = b - alpha * dj_db
        w = w - alpha * dj_dw

        if i<10000:
            J_history.append(cost_function(X, Y, w , b))
            p_history.append([w,b])
            
        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(iterations/10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
            
        return w,b,J_history
        
```
